```python
import pathlib
import os
import rich
import rich.table
import rich.markup
import pandas as pd


pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', 1000)
```

```python
path = pathlib.Path(f"{os.environ['SCRATCH']}/rejection_sampling_saves/gsm8k_5")

def extract_epoch(path):
    return int(path.parent.name.split("epoch_")[-1]) + 1

data = []
for run in path.glob("*"):
    checkpoints = list(run.glob("**/*.safetensors"))
    checkpoints.sort(key=lambda x: extract_epoch(x))
    max_epoch = extract_epoch(checkpoints[-1])
    data.append({
        "Run": run,
        "Checkpoint": len(list(run.glob("**/*.safetensors"))),
        "Creation Date": pd.Timestamp(run.stat().st_mtime, unit='s'),
        "Max Epoch": max_epoch,
        "Latest Checkpoint Creation Date": pd.Timestamp(checkpoints[-1].stat().st_mtime, unit='s')
    })
df = pd.DataFrame(data)
print(df)
df = df.sort_values("Latest Checkpoint Creation Date")
df

# # Get paths with 0 checkpoints
# qty = 1
# zero_checkpoint_paths = df[df['Checkpoint'] == qty]['Run'].tolist()

# # Delete the directories
# for path_to_delete in zero_checkpoint_paths:
#     try:
#         import shutil
#         detected_qty = len(list(path_to_delete.glob("**/*.safetensors")))
#         assert detected_qty == qty, f"Expected {qty} checkpoints, got {detected_qty}"
#         shutil.rmtree(path_to_delete)
#         print(f"Deleted: {path_to_delete}")
#     except Exception as e:
#         print(f"Error deleting {path}: {e}")

```
