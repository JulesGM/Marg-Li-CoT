```python
import datasets
import rich
import rich.panel
import rich.markup
import ast
import more_itertools as mit
import pathlib


def chat_box(role, role_input, role_style):
    # Find text between stars *adsasd* and add stars_style to it
    role_input = rich.markup.escape(role_input)

    panel = rich.panel.Panel(
        role_input, 
        title=f"{role_style}{role}[/]:", 
        title_align="left"
    )
    
    rich.print(panel)
```

```python
root = pathlib.Path("/home/mila/g/gagnonju/marglicot/with_open-instruct/light_eval_tests/outputs/details/")
ckpt = root / "_network_scratch_g_gagnonju_open_instruct_output_2024-12-31_21-52-49_rlvr_gsm8k_only_smollm2_instruct_checkpoints_step_700/2025-01-02T15-12-57.998700/details_custom|gsm8k|8_2025-01-02T15-12-57.998700.parquet"

details = datasets.load_dataset(
    "parquet", 
    data_files=[ckpt]
)
```

```python
def pyon_loads(s):
    return ast.literal_eval(s)
```

```python

def load_prediction(sample):
    text, logits = mit.one(pyon_loads(sample["predictions"]))
    return mit.one(text)

sample_index = 123
sample = details["train"][sample_index]
chat_box(role="Prompt", role_input=sample["full_prompt"]  , role_style="[bold blue]")
chat_box(role="Answer", role_input=load_prediction(sample), role_style="[bold blue]")
```
